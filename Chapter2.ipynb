**a.** The sample size *_n_* is extremely large, and the number of predictors *_p_* is small.

> We expect the more flexible statistical learning method to perform **better** because the method can better utilize the large number of observations to possibly find a more accurate shape of _f_.

**b.** The number of predictors *_p_* is extremely large, and the number of observations *_n_* is small.

> A more flexible method will perform **worse** because more flexible methods typically require a large number of observations in order to obtain an accurate estimate for _f_. Many predictors coupled with few observation typically leads to overfitting for more flexible methods.

**c.** The relationship between the predictors and response is highly non-linear.

> A more flexible method will perform **better** because more flexible methods can find non-linear shapes of _f_.

**d.** The variance of the error terms is extremely high.

> A more flexible method will perform **worse** because it is more likely to fit the noise (overfit) that find the true estimate of _f_.

**5. What are the advantages and disadvantages of a very flexible (versus a less flexible) approach for regression or classification? Under what circumstances might a more flexible approach be preferred to a less flexible approach? When might a less flexible approach be preferred?**

> The **advantages for a flexible approach** for regression or classification is that more flexible models can fit many different possible functional forms for *_f_*. Some of the **disadvantages for a more flexible approach** is that fitting a more flexible model requires estimating a greater number of parameters and that more complex (flexible) models can lead to a phenomenon known as _overfitting_ the data, which essentially means they follow the errors, or noise, too closely.
A more flexible approach may be preferred if you have no concern for interpretability, but instead only care about prediction. (But you need to be careful not to overfit!)
If we are mainly interested in inference, then less flexible models may be preferred as they are much more interpretable.

**6. Describe the differences between a parametric and a non-parametric statistical learning approach. What are the advantages of a parametric approach to regression or classification (as opposed to a non-parametric approach)? What are its disadvantages?**

> The **advantages of a parametric** approach to regression or classification are that it reduces the problem of estimating *_f_* down to one of estimating a set of parameters. This beneficial because it is generally much easier to estimte a set of parameters that it is to fit an entirely arbitrary fuction of *_f_*. The potential **disadvantage of a parametric** approach is that the model we choose will usually not match the true unknown form of *_f_*. If the chosen model is too far from the true *_f_*, then our estimate will be poor.
The **advantages of a non-parametric** approach is that by avoiding the assumption of a particular functional form for *_f_*, they have the potential to accurately fit a wider range of possible shapes for *_f_*. The **disadvantages of non-parametric** approaches is that since they do not reduce the problem of estimating *_f_* to a small number of parameters, a very large number of observations is required in order to obtain an accurate estimate for *_f_*.
